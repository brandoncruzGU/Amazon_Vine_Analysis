```python
#Import dependencies and file path.
import pandas as pd
file = "vine_table.csv"
```

```python
#Read the csv file.
file_csv = pd.read_csv(file)
```

```python
#Convert the csv file to a data frame.
vine_df = pd.DataFrame(file_csv)
vine_df
```

```python
#Filter the data to retrieve all the rows where total_votes count is greater than or equal to 20.
total_votes_df = vine_df[vine_df["total_votes"]>=20]
total_votes_df
```

```python
#Filter the data frame to retrieve the rows where the number of helpful votes divided by total votes is greater than or equal to 50%.
helpful_percentage = total_votes_df["helpful_votes"]/total_votes_df["total_votes"]
helpful_votes_df = total_votes_df[helpful_percentage>=.5]
helpful_votes_df
```

```python
#Filter the data to retrieve the rows where a review was written as part of the Vine paid program.
paid = (helpful_votes_df["vine"] == "Y")
paid_df = helpful_votes_df[paid == True]
paid_df
```

```python
#Filter the data to retrieve the rows where a review was not written as part of the Vine paid program.
unpaid = (helpful_votes_df["vine"] == "N")
unpaid_df = helpful_votes_df[unpaid == True]
unpaid_df
```

```python
#Calculate the total number of reviews for rows with Vine reviews.
total_vine_reviews = len(paid_df)
print(total_vine_reviews)
```

```python
#Calculate the total number of reviews for rows without Vine reviews.
total_nonvine_reviews = len(unpaid_df)
print(total_nonvine_reviews)
```

```python
#Calculate the number of 5-start reviews for rows with Vine reviews.
five_star_vines = sum(paid_df["star_rating"] == 5)
print(five_star_vines)
```

```python
#Calculate the number of 5-start reviews for rows without Vine reviews.
five_star_nonvines = sum(unpaid_df["star_rating"] == 5)
print(five_star_nonvines)
```

```python
#Calculate the percentage of 5-star ratings for Vine reviews.
vine_five_star_percent = five_star_vines/total_vine_reviews
print(vine_five_star_percent)
```

```python
#Calculate the percentage of 5-star ratings for non-Vine reviews.
nonvine_five_star_percent = five_star_nonvines/total_nonvine_reviews
print(nonvine_five_star_percent)
```
